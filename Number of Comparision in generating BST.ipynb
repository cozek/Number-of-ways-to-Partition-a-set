```python
#required for printing python dicts in a pretty way
import pprint

#required for generating permutations
import itertools
```

```python
# k = int(input("Enter a positive integer: "))

#hard coded for convenience
k = 4

l = [i for i in range(1,k+1)]

#generating all the possible permutations using itertools.permutations
bsts = list(itertools.permutations(l))
bsts = [list(i) for i in bsts]
bst_p = bsts.copy()
print("K: ",k)
print("Possible BSTS: ", len(bsts),"\n")

pp = pprint.PrettyPrinter(width = 70, compact= True)
pp.pprint(bst_p)
```

## Example of BST generation

```python
# let l be out given list of nodes
l = [4,1,2,3,6]

temp = l.copy() #save a copy for printing later

# the root node value, since we know that the root is at the 1st index
start = l[0]  

# init tree with the root node. 
# In our program , we use a python dictionary to store the tree
# Hence we need the root element stored at 'start' to reference the dict.
tree = { 
    
        l[0]: {'right': 'null', 'left': 'null'} 
       
       } 


# take a counter to store the number of comparisions
counter = 0  

# remove the 1st element(i.e the root element) from the given list
# we dont need it to be in l anymore as it is already in the tree and we 
# have 'start' to reference to it in the tree.
l.pop(0) 


# we create a loop that runs until l is empty, which can easily be checked 
# by counting the number of elements in it using the len() function.
while(len(l)!=0):
    
    # reset/set the pointer to point to the root , as we start our traversal 
    #  from the root each time while inserting a new node in the tree
    pointer = start
    
    # we pop the next time we want to insert into the tree from l and 
    # store it in 'current'. since we popped the number 
    # of elements in l will decrease by 1
    current = l.pop(0)
    
    # we start an infinite loop that breaks when we a node is 
    # successfully inserted into the tree. 
    # You can say this simulates a recursive behaviour without 
    # actually using recursion.
    while(1):
        
        # Now that we are in the infinite loop, we need to check if the 
        #'current' node is greater or smaller than node currently pointed
        #  to by the 'pointer' 
        
        
        if current < pointer:
            
            # a comparion was made, so we need to increment the counter by 1
            counter += 1
            
            # if our 'current' is lesser than the pointer node, then we 
            # need to insert it as the left child of the pointer node.
            # But we can only do that if 
            # it currently has no left child, i.e left child == NULL
            
            if tree[pointer]['left'] == 'null':
                # if the left child of pointer node is NULL, we update the 
                # status of the pointer node in the tree,i.e, 
                # set its left child as 'current'
                
                tree[pointer]['left'] = current
                
                # we then create 'current' as a new node and insert it 
                # into the tree
                new_node = {current:{'right': 'null', 'left':'null'}}
                tree.update(new_node)
                break
            
            if tree[pointer]['left'] != 'null':
                # if the left of pointer is not NULL, we need to 
                # compare current with the value of left child 
                # and then insert appropriately.
                # so we set the pointer to the left child and and  
                # go back to the start of the infinite loop
                
                pointer = tree[pointer]['left']
                continue
                
        elif current > pointer:
            
            # Works with the same logic as the preceding 'if' ,
            # if current is greater than the pointer, check if the right child
            # of the pointer is NULL. If its NULL , update the pointer node's 
            # right child as the 'current' and insert 'current' as a new node
            # and break out of the infinite loop.
            # If its not NULL, set the pointer to the right child and go back 
            # to the start of the infinite loop
            
            # a comparion was made, so we need to increment the counter by 1.
            counter += 1
            if tree[pointer]['right'] == 'null':
                tree[pointer]['right'] = current
                new_node = {current:{'right': 'null', 'left':'null'}}
                tree.update(new_node)
                break
            if tree[pointer]['right'] != 'null':
                pointer = tree[pointer]['right']
                continue
       
                

#pretty print the tree and print the number of comparisions made
print("Root node: ",temp[0])
pp.pprint(tree)
print("\n")
print("Comparisions made for ",temp, " : ", counter)




```

# Creating a function using the logic above to generate BSTs

```python
def genBST(l):
   
    temp = l.copy()

    start = l[0]  

    tree = { 

            l[0]: {'right': 'null', 'left': 'null'} 

           } 

    counter = 0  

    l.pop(0) 

    while(len(l)!=0):

        pointer = start

        current = l.pop(0)
        while(1):        

            if current < pointer:

                counter += 1
                if tree[pointer]['left'] == 'null':
                    tree[pointer]['left'] = current
                    new_node = {current:{'right': 'null', 'left':'null'}}
                    tree.update(new_node)
                    break

                if tree[pointer]['left'] != 'null':
                    pointer = tree[pointer]['left']
                    continue

            elif current > pointer:

                counter += 1
                if tree[pointer]['right'] == 'null':
                    tree[pointer]['right'] = current
                    new_node = {current:{'right': 'null', 'left':'null'}}
                    tree.update(new_node)
                    break
                if tree[pointer]['right'] != 'null':
                    pointer = tree[pointer]['right']
                    continue

    #print("Root node: ",temp[0])
    #pprint(tree)
    #print("\n")
    print("Comparisions made for ",temp, " : ", counter)
    
    return 



```
